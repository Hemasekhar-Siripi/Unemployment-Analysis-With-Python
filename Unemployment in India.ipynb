# PIE PLOTS

## According To The Area of"Unemployment in India"Dataset

```python
fig, ax = plt.subplots(figsize =(8, 7))
#labels
m=['Rural','Urban']
#title for chart
ax.set_title("\nACCORDING TO AREA IN \"Unemployment in india\" dataset",color="brown")
# Creating color parameters
colors = ( "blue" ,"magenta")
# Creating explode data
explode = (0.2, 0.0)
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
# Creating plot
wedges, texts, autotexts = ax.pie(ds2['Area'].value_counts(),
                                  autopct = "%0.2f",
                                  explode = explode,
                                  labels = m,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 105,
                                  wedgeprops = wp,
                                  textprops = dict(color ="brown"))
# Adding legend
ax.legend(wedges,m,
          title ="AREA",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")

plt.show()
```

## According to the Region Directions in"Unemployment Rate Upto 2020"Dataset

```python
ds1['Region.1'].unique()
```

```python
fig, ax = plt.subplots(figsize =(8, 7))
#labels
m=['South', 'Northeast', 'East', 'West', 'North']
#title for chart
ax.set_title("\nACCORDING TO REGION DIRECTIONS",color="brown")
# Creating color parameters
colors = ( "orange", "cyan", "brown","grey", "indigo")
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0)
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
# Creating plot
wedges, texts, autotexts = ax.pie(ds1['Region.1'].value_counts(),
                                  autopct = "%0.2f",
                                  explode = explode,
                                  labels = m,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
# Adding legend
ax.legend(wedges,m,
          title ="Directions",
          loc ="center left",
          bbox_to_anchor =(1.2, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")

plt.show()
```
